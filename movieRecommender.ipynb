```python
import numpy as np
import pandas as pd
```

```python
movies=pd.read_csv('archive/tmdb_5000_movies.csv')
credits=pd.read_csv('archive/tmdb_5000_credits.csv')
credits.head()
```

```python
# Merge the two dataset
movies=movies.merge(credits, on='title')
```

```python
movies.columns
```

```python
movies=movies[['movie_id', 'title','overview','genres', 'keywords','cast', 'crew']]
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
movies.isnull().sum()
```

```python
movies.duplicated().sum()
```

```python
import ast

def convert(obj):
  l=[]
  for i in ast.literal_eval(obj):
    l.append(i['name'])
  return l
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['keywords']= movies['keywords'].apply(convert)
```

```python
def convert3(obj):
  counter=0
  l=[]
  for i in ast.literal_eval(obj):
    l.append(i['name'])
    counter+=1
    if counter==3:
      break
  return l
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies['cast'][0]
```

```python
def fetchDirector(obj):
  l=[]
  for i in ast.literal_eval(obj):
    if i['job']=='Director':
      l.append(i['name'])
      break
  return l
```

```python
movies['crew']=movies['crew'].apply(fetchDirector)
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
def remove_space(item_list):
    return [i.replace(" ", "") for i in item_list]

cols_to_fix = ['cast', 'crew', 'genres', 'keywords']

for col in cols_to_fix:
    movies[col] = movies[col].apply(remove_space)
```

```python
movies.head()
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
new_df=movies[['movie_id','title','tags']]
new_df.head()
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()


def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
new_df['tags']=new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
def recommend(movie):
  movie_index=new_df[new_df['title']==movie].index[0]
  distances=similarity[movie_index]
  movies_list=sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]
  for i in movies_list:
    print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```

```python
import pickle
pickle.dump(new_df,open('movie_list.pkl','wb'))
pickle.dump(similarity,open('similarity.pkl','wb'))
```
